```python
# matlabpython.

import cv2
import numpy as np
import matplotlib.pyplot as plt

def fenduanxianxing(image):
    height, width = image.shape[:2]
    dst = np.zeros([height, width])
    for x in range(height):
        for y in range(width):
            if image[x,y]<150:
                dst[x,y] = 1.25*image[x,y]
            else:
                dst[x,y] = 0.523*image[x,y] + 176
    return dst

def feixianxing(image):
    height, width = image.shape[:2]
    dst = np.zeros([height, width])
    for x in range(height):
        for y in range(width):
          dst[x,y] = 1/255*image[x,y]*image[x,y]  
    return dst

img = cv2.imread('touxiang.jpg', 0)

img1 = fenduanxianxing(img)
img2 = feixianxing(img)

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(img1, 'gray')
plt.subplot(223), plt.imshow(img2, 'gray')
plt.show()
```
